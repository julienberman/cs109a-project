```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/cleaned_wvs_data.csv')
```

```python
fig, ax = plt.subplots(2, 2, figsize=(12, 8))
ax = ax.ravel() 

x_vars = ['work_ethic', 'trust', 'equality', 'religion']
titles = ['Democracy Index vs. Work Ethic', 'Democracy Index vs. Trust', 
          'Democracy Index vs. Equality', 'Democracy Index vs. Religion']

for i, (x_var, title) in enumerate(zip(x_vars, titles)):
    sns.scatterplot(x=x_var, y='democracy_score', data=df, alpha=0.7, ax=ax[i])
    ax[i].set_title(title)
    ax[i].set_xlabel(x_var.replace('_', ' ').title()) 
    ax[i].set_ylabel('Democracy Index')
    ax[i].grid()

plt.tight_layout()
plt.show();
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 8))
ax = ax.ravel() 

x_vars = ['gdp_per_capita', 'hdi']
titles = ['Democracy Index vs. GDP per Capita', 'Democracy Index vs. HDI']

for i, (x_var, title) in enumerate(zip(x_vars, titles)):
    sns.scatterplot(x=x_var, y='democracy_score', data=df, alpha=0.7, ax=ax[i])
    ax[i].set_title(title)
    ax[i].set_xlabel(x_var.replace('_', ' ').title())
    ax[i].set_ylabel('Democracy Index')
    ax[i].grid()

plt.tight_layout()
plt.show()
```
